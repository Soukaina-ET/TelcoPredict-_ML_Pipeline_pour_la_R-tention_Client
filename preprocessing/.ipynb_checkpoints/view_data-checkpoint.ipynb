Part 1: Statistiques Descriptives (Mesures de tendance centrale et dispersion)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv(
    r"C:\Users\lalib\Documents\DL\doc\data\WA_Fn-UseC_-Telco-Customer-Churn.csv"
)

# 1. Mesures statistiques pour les variables numériques
# On inclut l'asymétrie (skewness) et l'aplatissement (kurtosis)
stats = df.describe().T
stats['skew'] = df.skew(numeric_only=True)
stats['kurtosis'] = df.kurtosis(numeric_only=True)
print(stats)
```

Part 2: Visualisations Univariées (Comprendre chaque variable)

```python
# Configuration du style
sns.set_theme(style="whitegrid")
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Distribution de l'ancienneté (Tenure)
sns.histplot(df['tenure'], kde=True, ax=axes[0], color='skyblue')
axes[0].set_title('Distribution de l\'Ancienneté (Mois)')

# Distribution des charges mensuelles
sns.histplot(df['MonthlyCharges'], kde=True, ax=axes[1], color='salmon')
axes[1].set_title('Distribution des Charges Mensuelles')

# Répartition du Churn (Variable Cible)
df['Churn'].value_counts().plot.pie(autopct='%1.1f%%', ax=axes[2], colors=['#66b3ff','#99ff99'])
axes[2].set_title('Répartition du Churn (Déséquilibre de classe)')

plt.show()
```

Part 3: Visualisations Bivariées (Le lien avec le Churn)

```python
# Quel est l'impact du type de contrat sur le départ ?
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='Contract', hue='Churn')
plt.title('Impact du Contrat sur le Churn')
plt.show()

# Boxplot : Les clients qui partent paient-ils plus cher par mois ?
plt.figure(figsize=(8, 5))
sns.boxplot(x='Churn', y='MonthlyCharges', data=df)
plt.title('Charges Mensuelles vs Churn')
plt.show()
```
